# In this script, time series of puckering parameters and dihedral will be analysed

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def unwrap(raw):
    con = np.copy(raw)
    for i in range(1,raw.size): 
        diff=con[i]-con[i-1]
        if diff > 180:
            con[i:] -= 360
        elif diff < -180:
            con[i:] += 360
    return con

def myplot(name):
    rawdis = np.genfromtxt("/home/winter/onsager/"+name+".dis", skip_header=1, skip_footer=0,)
    sta = np.genfromtxt("/home/winter/onsager/"+name+".sta", skip_header=1, skip_footer=0, usecols=0)
    rot = np.genfromtxt("/home/winter/onsager/"+name+".rot", skip_header=1, skip_footer=0)
    dis = unwrap(rawdis)
    plt.figure(figsize=(5,4))
    plt.plot(dis)
    plt.plot(sta,color='black')
    plt.plot(rot)
    plt.xlabel('Trajectory Frame No.')
    plt.ylabel('Value of Dihedral or Pucker (degree)')
    plt.xlim(257500,262500)
    plt.ylim(-200,300)
    plt.grid(True,axis='y')
    plt.legend(['central torsion', 'stator pucker', 'rotor pucker'], loc='lower right')
    plt.savefig('/home/winter/onsager/Pucker-nearTHI.png', dpi=1200, bbox_inches = 'tight')
    plt.show()
    

name = "amber/Snolog2"
myplot(name)
#myplot2("amber/test"+str(i))
```
